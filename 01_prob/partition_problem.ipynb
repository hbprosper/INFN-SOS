$\color{blue}{n_S = 15}$ students, 
$\color{blue}{n_U = 12}$ undergraduates and $\color{blue}{n_G = 3}$  graduate students are randomly divided into $\color{blue}{n_G = 3}$ groups of $\color{blue}{n_g = 5}$ students each [1]. What is the probability that each group includes a single graduate student?

We use the Python module **itertools** to simulate the problem.

```python
# combinatorial analysis
from itertools import combinations
from math import factorial
```

## Compute total number of partitions
We divide the $n_S=15$ students into $n_G=3$ groups of $n_g = 5$ students per group. Since there are $n_S$ students to fill $n_S$ positions within the groups, the students can be arranged among these positions in $n_S!$ ways. But, since the order of students within a group is irrelevant, for each group the number of partitions is too large by $n_g!$. Therefore, the cardinality of the set $\color{blue}{\Omega}$ of partitions is
\begin{align}
|\Omega| & = \frac{n_S !}{(n_g!)^{n_G}},\\
    & = \frac{15!}{(5!)^3} = 756,756.
\end{align}

Here is the cardinality of the set for the problem posed in the Lecture 1.

```python
nS = 16
nG =  4
OMG = factorial(16)/(factorial(4))**4
print(f'{OMG:,}')
```

```python
nS = 15          # number of students
nG =  3          # number of grad students (same as the number of groups)
ng =int(nS / nG) # number of students per group
nU = nS-nG       # number of undergrads

U = ['U%2.2d' % x for x in range(1, nU+1)] # undergrads
G = ['G%2.2d' % x for x in range(1, nG+1)] # grads
S = set(U + G)

def printit():
    print(f'''
    number of students (nS):                {nS:5d}
    number of groups (grad) students (nG):  {nG:5d}
    number of undergrads (nU):              {nU:5d}
    number of students/group (ng):          {ng:5d}
    set of students:\nS: {str(S):s}
    ''')
printit()

M = int(factorial(nS)/factorial(ng)**nG)
print(f'|Omega|: {M:10,d}')
```

## Count number of partitions 
The code below seems to work only if the $n$ in $\binom{m}{n}$ differs between the two calls to `combinations`.

```python
Omega = [] # Sample space of partitions

# C1, C2, C3 are the sets of students assigned to groups 1, 2, and 3, respectively.
for C1 in combinations(S, ng):
    C1 = set(C1)
    # Remove students assigned to group 1 from the set S of all students
    S1 = S - C1

    # Get students for group 2
    for C2 in combinations(S1, ng):
        C2 = set(C2)
        # Remove students assigned to group 2 from the set S1
        # The remaining students are assigned to group 3
        C3 = S1 - C2

        # Save partition
        Omega.append([C1, C2, C3])

print(f'|Omega|: {len(Omega):10,d}')
```

## Count number of partitions with 1 graduate student per group
Since we want 1 graduate per group of size $n_g$, there remain $n_g - 1$ positions per group available for undergraduates. Given that there are $n_U$ undergraduates for $n_U$ positions, the number of partitions of undergraduates is, therefore,
\begin{align}
K & = \frac{n_U !}{((n_g-1)!)^{n_G}},\\
    & = \frac{12!}{(4!)^3} = 34,650.
\end{align}
But for every one of these partitions of undergraduates the grad students can be arranged among themselves $n_G!$ ways. Therefore, the cardinality of the set of the desired experimental outcomes, $E$, is
\begin{align}
|E| & = n_G! \frac{n_U !}{((n_g-1)!)^{n_G}},\\
    & = 3! \frac{12!}{(4!)^3} = 207,900.
\end{align}

```python
print(f'Number of undergrads:                 {nU:10d}')
print(f'Number of undergraduates/group:       {ng-1:10d}')
print(f'Number of groups:                     {nG:10d}')

K = int(factorial(nU)/factorial(ng-1)**nG)
print(f'number of combinations of undergrads: {K:10,d}')

E = int(factorial(nG)) * K
print(f'|E|: {E:10,d}')
```

## Explicit construction of the set $E$

Let $G$ denote the set of graduate students and, for a given partition $x$, let $C_1$, $C_2$, and $C_3$ be the sets of students in each group. Here is one way to define the set $E$:
\begin{align*}
    E & = \{ x : (|G\cap C_1| == 1) \text{ and } (|G\cap C_2| == 1) \text{ and } (|G\cap C_3| == 1) \}.
\end{align*}

```python
G = set(G)
print(f'Set of graduate students: {G}')

# Loop over the sample space
K = 0
for C1, C2, C3 in Omega:

    if len(G.intersection(C1)) != 1:
        continue
        
    if len(G.intersection(C2)) != 1: 
        continue
        
    if len(G.intersection(C3)) != 1:
        continue

    # By construction, each partition contains single graduate student
    K += 1

print(f'|E|: {K:10,d}')
```
